- Ê†πÊçÆÊâßË°åÁªìÊûúÔºåÂΩìÂâçÁõÆÂΩï‰∏ãÂ≠òÂú®ÁöÑÊòØ`test.dot`Êñá‰ª∂ÔºåÂõ†Ê≠§‰øÆÊîπ‰∫ÜÊñá‰ª∂Ë∑ØÂæÑ
- Â¶ÇÊûúÁéØÂ¢É‰∏≠Ê≤°ÊúâÂÆâË£Ö`pydot`ÔºåÂèØËÉΩ‰ºöÊä•ÈîôÔºåÈúÄË¶ÅÂÖàÂÆâË£ÖÔºö`!pip install pydot`
- ‰ª£Á†Å‰ºöÊ≠£Â∏∏Ê£ÄÊü•Êñá‰ª∂ÊòØÂê¶Â≠òÂú®ÔºåÂ¶ÇÊûú‰∏çÂ≠òÂú®‰ºöÊäõÂá∫`FileNotFoundError`ÂºÇÂ∏∏

```python
# Ê≠•È™§ 2: ËØªÂèñÊú¨Âú∞ÁöÑdotÊñá‰ª∂
import os
import networkx as nx

# Ê£ÄÊü•ÂΩìÂâçÂ∑•‰ΩúÁõÆÂΩïÂíåÂèØÁî®Êñá‰ª∂
print("ÂΩìÂâçÂ∑•‰ΩúÁõÆÂΩï:", os.getcwd())
print("ÂΩìÂâçÁõÆÂΩï‰∏ãÁöÑÊñá‰ª∂:", [f for f in os.listdir('.') if f.endswith('.dot')])

# ‰ΩøÁî®read_dotÂáΩÊï∞ËØªÂèñdotÊñá‰ª∂
# Â¶ÇÊûúÊñá‰ª∂‰∏çÂ≠òÂú®ÔºåÂ∞ÜÊ≠£Â∏∏Êä•Èîô
G = nx.drawing.nx_pydot.read_dot("test.dot")  # ËØªÂèñÊñá‰ª∂ÔºåÁªìÊûúÂ≠òÂÇ®‰∏∫networkxÂõæÂΩ¢ÂØπË±°G

# ÊâìÂç∞ÂõæÂΩ¢Âü∫Êú¨‰ø°ÊÅØ
print(f"ÊàêÂäüËØªÂèñÂõæÂΩ¢ÔºåÂåÖÂê´ {G.number_of_nodes()} ‰∏™ËäÇÁÇπÂíå {G.number_of_edges()} Êù°Ëæπ")
```

**ÊèèËø∞**: ‰ΩøÁî®networkxÁöÑÁªòÂõæÂáΩÊï∞ÔºàÂ¶ÇdrawÔºâÁªìÂêàmatplotlibÊòæÁ§∫ÂõæÂΩ¢„ÄÇÂèØ‰ª•Ëá™ÂÆö‰πâÂ∏ÉÂ±ÄÔºàÂ¶Çspring_layoutÔºâÂíåÊ†∑ÂºèÔºåÁÑ∂ÂêéË∞ÉÁî®matplotlibÁöÑshowÂáΩÊï∞Êù•Â±ïÁ§∫ÂèØËßÜÂåñÁªìÊûú„ÄÇ

**È¢ÑÊúüËæìÂá∫**: ‰∏Ä‰∏™ÂõæÂΩ¢Á™óÂè£ÊàñÂõæÂÉèÔºåÊ∏ÖÊô∞Â±ïÁ§∫dotÊñá‰ª∂‰∏≠ÁöÑÁΩëÁªúÁªìÊûÑÔºåÂåÖÊã¨ËäÇÁÇπÂíåËøûÊé•ÂÖ≥Á≥ª„ÄÇ

## ÂèØËßÜÂåñÂõæÂΩ¢‰ª£Á†ÅËØ¥Êòé

Âü∫‰∫éÂâçÈù¢Â∑≤ÂÆåÊàêÁöÑÊ≠•È™§ÔºåÁé∞Âú®ÈúÄË¶ÅÂ∞ÜËØªÂèñÁöÑÂõæÂΩ¢ËøõË°åÂèØËßÜÂåñ„ÄÇ‰ª£Á†ÅÂ∞ÜÔºö
1. ‰ΩøÁî®matplotlibÂàõÂª∫ÂõæÂΩ¢Á™óÂè£
2. Â∫îÁî®spring_layoutÂ∏ÉÂ±ÄÁÆóÊ≥ï
3. ‰ΩøÁî®networkxÁöÑÁªòÂõæÂáΩÊï∞ÁªòÂà∂ÂõæÂΩ¢
4. Ëá™ÂÆö‰πâËäÇÁÇπÈ¢úËâ≤„ÄÅÂ§ßÂ∞èÂíåÊ†áÁ≠æÊ†∑Âºè
5. ÊòæÁ§∫ÊúÄÁªàÁöÑÂèØËßÜÂåñÁªìÊûú

ËøôÊÆµ‰ª£Á†ÅÂ∞Ü‰ΩøÁî®ÂâçÈù¢Ê≠•È™§‰∏≠ËØªÂèñÁöÑÂõæÂΩ¢ÂØπË±° `G`ÔºåÈÄöËøánetworkxÂíåmatplotlibËøõË°åÂèØËßÜÂåñÂ±ïÁ§∫„ÄÇ‰ª£Á†ÅÂåÖÂê´‰∫ÜÂ§öÁßçÊ†∑ÂºèËá™ÂÆö‰πâÈÄâÈ°πÔºåÂ¶ÇËäÇÁÇπÈ¢úËâ≤„ÄÅÂ§ßÂ∞è„ÄÅÊ†áÁ≠æÊ†∑ÂºèÁ≠âÔºå‰ΩøÂõæÂΩ¢Êõ¥Âä†Ê∏ÖÊô∞ÊòìËØª„ÄÇ

```python
# Ê≠•È™§ 3: ÂèØËßÜÂåñÂõæÂΩ¢
import matplotlib.pyplot as plt

# ËÆæÁΩÆÂõæÂΩ¢Â§ßÂ∞èÂíåÂ∏ÉÂ±Ä
plt.figure(figsize=(10, 8))

# ‰ΩøÁî®spring_layoutËÆ°ÁÆóËäÇÁÇπ‰ΩçÁΩÆ
pos = nx.spring_layout(G)

# ÁªòÂà∂ÂõæÂΩ¢ÔºåËá™ÂÆö‰πâÊ†∑Âºè
nx.draw(G, pos, 
        with_labels=True,      # ÊòæÁ§∫ËäÇÁÇπÊ†áÁ≠æ
        node_color='lightblue', # ËäÇÁÇπÈ¢úËâ≤
        node_size=800,         # ËäÇÁÇπÂ§ßÂ∞è
        font_size=12,          # Ê†áÁ≠æÂ≠ó‰ΩìÂ§ßÂ∞è
        font_weight='bold',    # Ê†áÁ≠æÂ≠ó‰ΩìÁ≤óÁªÜ
        edge_color='gray',     # ËæπÈ¢úËâ≤
        width=2,               # ËæπÂÆΩÂ∫¶
        alpha=0.7)             # ÈÄèÊòéÂ∫¶

# ËÆæÁΩÆÂõæÂΩ¢Ê†áÈ¢ò
plt.title("Network Graph Visualization from DOT File", fontsize=16, pad=20)

# ÁßªÈô§ÂùêÊ†áËΩ¥
plt.axis('off')

# ÊòæÁ§∫ÂõæÂΩ¢
plt.tight_layout()
plt.show()

# ÊâìÂç∞ÂèØËßÜÂåñÂÆåÊàê‰ø°ÊÅØ
print("ÂõæÂΩ¢ÂèØËßÜÂåñÂÆåÊàêÔºÅ")
```
